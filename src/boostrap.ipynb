```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
diabetes_df = pd.read_csv('/content/clean_data/diabetes.csv')
```

```python
# here is the sample of 25 observations
sample_size = 25
sample_data = diabetes_df['Glucose'].sample(n=sample_size, random_state=42)
```

```python
# Now we can compare with the barcharts
labels = ['Population Mean', 'Population Highest', 'Sample Mean', 'Sample Highest']
values = [p_mean, h_glucose, s_mean, hi_glucose]

plt.bar(labels, values, color=['blue', 'blue', 'orange', 'orange'])
plt.title('Comparison of Glucose Statistics')
plt.ylabel('Glucose Values')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Setting the seed value
np.random.seed(25)
```

```python
# here is the data set
diabetes_df = pd.read_csv('/content/clean_data/diabetes.csv')
```

```python
# now we have to extract the bloodpressure  column from the diabetes dataset
p_blood_pressure = diabetes_df['BloodPressure']
```

```python
# Now we have to perform the bootstrap
def bootstrap_s(data, num_samples=500, sample_size=150):
    return [np.random.choice(data, size=sample_size, replace=True) for _ in range(num_samples)]
```

```python
# Performing the bootstrap sampling
b_samples = bootstrap_s(p_blood_pressure)
```

```python

# Calculation of bootstrap and population samples
pop_mean = np.mean(p_blood_pressure)
pop_std_dev = np.std(p_blood_pressure)
pop_percentile = np.percentile(p_blood_pressure, 95)
boo_means = np.mean(b_samples, axis=1)
boo_std_devs = np.std(b_samples, axis=1)
boo_percentiles = np.percentile(b_samples, 95, axis=1)


```

```python
# Here are the comparsion charts
plt.figure(figsize=(12, 4))

# comparsion of mean data
plt.subplot(1, 3, 1)
plt.hist(boo_means, bins=30, color='blue', alpha=0.7, label='Bootstrap Means')
plt.axvline(pop_mean, color='red', linestyle='dashed', linewidth=2, label='Population Mean')
plt.title('Mean ')
plt.legend()

```

```python

# Std Dev comparison
plt.subplot(1, 3, 2)
plt.hist(boo_std_devs, bins=30, color='orange', alpha=0.7, label='Bootstrap Std Devs')
plt.axvline(pop_std_dev, color='red', linestyle='dashed', linewidth=2, label='Population Std Dev')
plt.title('Std Dev')
plt.legend()

```

```python
# Percentile comparison
plt.subplot(1, 3, 3)
plt.hist(boo_percentiles, bins=30, color='green', alpha=0.7, label='Bootstrap 95th Percentiles')
plt.axvline(pop_percentile, color='red', linestyle='dashed', linewidth=2, label='Population 95th Percentile')
plt.title('Percentile ')
plt.legend()
```

```python
plt.tight_layout()
plt.show()
```
