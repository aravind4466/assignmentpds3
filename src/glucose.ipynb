```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
np.random.seed(25)
```

```python
diabetes_df = pd.read_csv('/content/clean_data/diabetes.csv')
```

```python
# here is the sample of 25 observations
sample_size = 25
sample_data = diabetes_df['Glucose'].sample(n=sample_size, random_state=42)
```

```python
# here we are finding the mean and highest glucose values
p_mean = diabetes_df['Glucose'].mean()
h_glucose = diabetes_df['Glucose'].max()

s_mean = sample_data.mean()
hi_glucose = sample_data.max()
```

```python
# here are the final results after calculation of mean and highest values
print(f"Mean Glucose: {p_mean}")
print(f"Highest Glucose: {h_glucose}")

print(f"Sample Glucose: {s_mean}")
print(f"Sample Hi Glucose: {hi_glucose}")


```

```python
# Now we can compare with the barcharts
labels = ['Population Mean', 'Population Highest', 'Sample Mean', 'Sample Highest']
values = [p_mean, h_glucose, s_mean, hi_glucose]

plt.bar(labels, values, color=['blue', 'blue', 'orange', 'orange'])
plt.title('Comparison of Glucose Statistics')
plt.ylabel('Glucose Values')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# we are setting the seed value
np.random.seed(25)
```

```python

# the dataset is uploaded
diabetes_df = pd.read_csv('/content/clean_data/diabetes.csv')
```

```python
#Here we are getting the BMI
sample_data = diabetes_df['BMI'].sample(n=sample_size, random_state=42)
```

```python
# Calculation of 98th percentile
p_98th_percentile = np.percentile(diabetes_df['BMI'], 98)
s_98th_percentile = np.percentile(sample_data, 98)
```

```python
# bars for showing comparsion
labels = ['Population', 'Sample']
values = [p_98th_percentile, s_98th_percentile]

plt.bar(labels, values, color=['green', 'red'])
plt.title('Comparison of 98th Percentile BMI')
plt.ylabel('BMI Values')
plt.show()
```
